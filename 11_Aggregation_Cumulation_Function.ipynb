# **Python - Cumulative Function**

```python
df = pd.DataFrame({
    "A" : [34,56,38,85,38],
    "B" : [46,27,59,37,49],
    "C" : [30,48,20,57,39],
    "D" : [33,47,93,25,83]
})
df
```

```python
# cumulative sum of values over the index axis
df.cumsum(axis = 0)
```

```python
# cumulative sum along column axis
df.cumsum(axis = 1)
```

```python
df = pd.DataFrame({
    "A" : [34,56,38,None,38],
    "B" : [46,None,59,37,49],
    "C" : [30,48,20,57,None],
    "D" : [None,47,93,25,83]
})
df
```

```python
# Cumulative Sum
df.cumsum(axis = 0, skipna = True)
```

```python
df = pd.DataFrame({
    "A" : [34,56,38,85,38],
    "B" : [46,27,59,37,49],
    "C" : [30,48,20,57,39],
    "D" : [33,47,93,25,83]
})
df
```

```python
# cumprod() - Return cumulative product over a DataFrame or Series axis
df.cumprod(axis = 0)
```

```python
#  cummax() - Return cumulative maximum over a DataFrame or Series axis.
df.cummax(axis = 0)
```

```python
# cummin() - Return cumulative minimum over a DataFrame or Series axis.
df.cummin(axis = 0)
```
