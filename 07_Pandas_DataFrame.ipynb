# **Pandas Data Frame**

```python
import pandas as pd
```

```python
# Creating a Data Frame using List
s1 = ['Python', 'Petroleum', ' Porosity', 'Drilling', 'Reservoir', 'Permeability', 'Saturation', 'Wettability']
d1 = pd.DataFrame(s1)
d1
```

```python
# Creating a Data Frame using List of Dictionary
dict1 = dict({'keywords' : ['Python', 'Petroleum', ' Porosity', 'Drilling', 'Reservoir', 'Permeability', 'Saturation', 'Wettability'],
              'assigned_values' : [12342,34632,3452423,47624,4632827,27463,234623,432746]})
d2 = pd.DataFrame(dict1)
d2
```

```python
# Creating a DataFrame using dict of Numpy Array
import numpy as np
d1 = {
    "Name" : np.array(['Alice', "Bob", 'Jack', "Neu", "Rose"]),
    "Age" : np.array([23,31,27,25,22]),
    "Salary" : np.array([50000,65000,70000,55000,60000])
}
d2 = pd.DataFrame(d1)
d2
```

```python
# Example of Petroleum Data
Capex = pd.DataFrame({
    "Drilling_CAPEX" : [2000000,2200000,2100000],
    'Completion_Capex' : [5000000,5200000,5150000],
    "Production_CAPEX" : [500000,550000,450000]
}, index = ['Well_1', 'Well_2', 'Well_3'])
Capex
```
