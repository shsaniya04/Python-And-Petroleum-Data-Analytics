# **Python - Descriptive Statics Basics**

```python
import pandas as pd
import numpy as  np
```

```python
df = pd.Series([1,2,3,4,5])
df
```

```python
df.describe()
```

```python
df_1 = pd.Series(['a', 'a', 'b', 'c', 'd', 'e', 'f'])
df_1
```

```python
df_1.describe()
```

```python
df = pd.DataFrame({
    "name": ["Alice", "Bob", "Charlie"],
    "age": [25, 30, 35],
    "city": ["New York", "San Francisco", "Los Angeles"]
})
df
```

```python
df.describe()
```

```python
df.describe(include = "all")
```

```python
df.age.describe()
```

```python
# Including only numeric columns in DataFrame Description
df.describe(include = np.number)
```

```python
# Including only string columns in DataFrame Description
df.describe(include = np.object_)
```

```python
df.describe(exclude = np.number)
```

```python
df.describe(exclude = [object])
```
